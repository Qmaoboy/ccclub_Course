```python
a="statue,stand,student"
b="dog,cat,does"
c="outstanding,asdasds,outside"
d=",,sdaddad"
e="apple,append,application,applause"
def p4(a):
    a=a.split(",")
    res=""
    vomin=float("inf")
    t=0
    for i in a:
        if len(i)<=vomin:
            vomin=len(i)
            t=i
    for idx in range(vomin):
        for i in a:
            if i[idx]!=t[idx]:
                return res
        res+=t[idx]
    return res
res=p4(c)
if res:
    print( res)
else:
    print("No common")
```

# 串列切割

```python
cin =["1,2,3","1,2,3,4","2,-2,1,-1,0,0","1,2,4,5,6,76",""]

def p5(cin):
    if not cin:
        return False
    cin=list(map(float,cin.split(",")))
    idx=0
    for i in range(len(cin)):
        if sum(cin[:idx])==sum(cin[idx:]):
            return True
        else:
            idx+=1
    return False

for i in cin:
    print(p5(i))
```
